```python
import numpy as np
import json 

gates = ["c8", "c9", "c10", "c11"] 
measurements = np.linspace(0, 1500, 16)
```

```python
for gate in gates:
    
    for measurement in measurements:
        
        config_file = 'ist1_crosstalk_config.json'
        
        print("setting: ", gate, measurement)
        pygor.setval(gate, measurement)
        
        
        with open(config_file) as f: 
            data = json.load(f)
        
        data["save_dir"] = gate + "_" + str(measurement) + "_" + data["save_dir"] 
        
        new_save_name = gate + str(measurement) + config_file
        
        with open(new_save_name, "w") as f:
            json.dump(data, f)
        
        #print("tune_with_pygor((newsavename))")
        a = tune_with_pygor_from_file(new_save_name)
        del a
        
    pygor.setval(gate, 0)
```
